```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
simulation = 'store-with-covid'
df = pd.read_csv(f'./out/{simulation}.csv', index_col=0)
df.head()
```

```python
fig, axs = plt.subplots(1,len(df.columns), figsize=(15, 6), facecolor='w', edgecolor='k')
fig.subplots_adjust(hspace = .5, wspace=.001)
axs = axs.ravel()
for i in range(len(df.columns)):
    col = df.columns[i]
    axs[i].plot(df.index, df[col])
    axs[i].set_title(col)
```
